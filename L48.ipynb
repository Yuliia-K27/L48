Застосуйте описані в лекції методи траснформації даних на датасеті Data_to_Transform, який містить:
    
-помірно зміщені праворуч дані (Moderate Positive Skew);
-сильно зміщені праворуч дані (Highly Positive Skew);
-помірно зміщені ліворуч дані (Moderate Negative Skew);
-сильно зміщені ліворуч дані (Highly Negative Skew).

Оцініть результати трансформацій через квантильно-квантильні графіки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import probplot

# Завантаження датасету
data_transf = pd.read_csv('Data_to_Transform.csv')
data_transf
```

```python
data_transf.columns
```

```python
# Логарифмічна трансформація
data_transf['Moderate_Positive_Skew_log'] = np.log(data_transf['Moderate Positive Skew'])
data_transf['Highly_Positive_Skew_log'] = np.log(data_transf['Highly Positive Skew'])
data_transf['Moderate_Negative_Skew_log'] = np.log(data_transf['Moderate Negative Skew'])
data_transf['Highly_Negative_Skew_log'] = np.log(data_transf['Highly Negative Skew'])

# Візуалізація
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Логарифмічна трансформація')

sns.histplot(data_transf['Moderate_Positive_Skew_log'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (log transformed)')

sns.histplot(data_transf['Highly_Positive_Skew_log'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (log transformed)')

sns.histplot(data_transf['Moderate_Negative_Skew_log'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (log transformed)')

sns.histplot(data_transf['Highly_Negative_Skew_log'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew (log transformed)')

plt.tight_layout()
plt.show()
```

```python
# Квантильно-квантильні графіки
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('QQ-графіки для логарифмічної трансформації')

probplot(data_transf['Moderate_Positive_Skew_log'], plot=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (log transformed)')

probplot(data_transf['Highly_Positive_Skew_log'], plot=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (log transformed)')

probplot(data_transf['Moderate_Negative_Skew_log'], plot=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (log transformed)')

probplot(data_transf['Highly_Negative_Skew_log'], plot=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew (log transformed)')

plt.tight_layout()
plt.show()
```

```python
# Трансформація кубічним коренем
data_transf['Moderate_Positive_Skew_cbrt'] = np.cbrt(data_transf['Moderate Positive Skew'])
data_transf['Highly_Positive_Skew_cbrt'] = np.cbrt(data_transf['Highly Positive Skew'])
data_transf['Moderate_Negative_Skew_cbrt'] = np.cbrt(data_transf['Moderate Negative Skew'])
data_transf['Highly_Negative_Skew_cbrt'] = np.cbrt(data_transf['Highly Negative Skew'])

# Візуалізація
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Трансформація кубічним коренем')

sns.histplot(data_transf['Moderate_Positive_Skew_cbrt'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (cbrt transformed)')

sns.histplot(data_transf['Highly_Positive_Skew_cbrt'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (cbrt transformed)')

sns.histplot(data_transf['Moderate_Negative_Skew_cbrt'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (cbrt transformed)')

sns.histplot(data_transf['Highly_Negative_Skew_cbrt'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew(cbrt transformed)')

plt.tight_layout()
plt.show()
```

```python
# Квантильно-квантильні графіки
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('QQ-графіки для трансформації кубічним коренем')

probplot(data_transf['Moderate_Positive_Skew_cbrt'], plot=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (cbrt transformed)')

probplot(data_transf['Highly_Positive_Skew_cbrt'], plot=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (cbrt transformed)')

probplot(data_transf['Moderate_Negative_Skew_cbrt'], plot=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (cbrt transformed)')

probplot(data_transf['Highly_Negative_Skew_cbrt'], plot=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew (cbrt transformed)')

plt.tight_layout()
plt.show()
```

```python
# Трансформація бокса-кокса
# Так як цей метод працює тільки з позитивними значеннями, використаємо альтернативу Йео-Джонсона

# Трансформація бокса-кокса з альтернативою Йео-Джонсона
data_transf['Moderate_Positive_Skew_boxcox'], _ = stats.yeojohnson(data_transf['Moderate Positive Skew'])
data_transf['Highly_Positive_Skew_boxcox'], _ = stats.yeojohnson(data_transf['Highly Positive Skew'])
data_transf['Moderate_Negative_Skew_boxcox'], _ = stats.yeojohnson(data_transf['Moderate Negative Skew'])
data_transf['Highly_Negative_Skew_boxcox'], _ = stats.yeojohnson(data_transf['Highly Negative Skew'])

# Візуалізація
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Трансформація бокса-кокса (Йео-Джонсон)')

sns.histplot(data_transf['Moderate_Positive_Skew_boxcox'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (Box-Cox transformed)')

sns.histplot(data_transf['Highly_Positive_Skew_boxcox'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (Box-Cox transformed)')

sns.histplot(data_transf['Moderate_Negative_Skew_boxcox'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (Box-Cox transformed)')

sns.histplot(data_transf['Highly_Negative_Skew_boxcox'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew (Box-Cox transformed)')

plt.tight_layout()
plt.show()
```

```python
# Квантильно-квантильні графіки

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Квантильно-квантильні графіки для трансформації бокса-кокса')

stats.probplot(data_transf['Moderate_Positive_Skew_boxcox'], plot=axes[0, 0])
axes[0, 0].set_title('Moderate Positive Skew (Box-Cox transformed)')

stats.probplot(data_transf['Highly_Positive_Skew_boxcox'], plot=axes[0, 1])
axes[0, 1].set_title('Highly Positive Skew (Box-Cox transformed)')

stats.probplot(data_transf['Moderate_Negative_Skew_boxcox'], plot=axes[1, 0])
axes[1, 0].set_title('Moderate Negative Skew (Box-Cox transformed)')

stats.probplot(data_transf['Highly_Negative_Skew_boxcox'], plot=axes[1, 1])
axes[1, 1].set_title('Highly Negative Skew (Box-Cox transformed)')

plt.tight_layout()
plt.show()
```
